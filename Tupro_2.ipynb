```python
def plot_linguistik(batas, warna, label='', min=0, max=1):
    y = [0, 1, 1, 0]
    plt.plot( (min, batas[0]), (0,0), warna )                                
    for i in range(len(batas)-1):                                             
      plt.plot( (batas[i], batas[i+1],), (y[i], y[i+1]), warna )
    plt.plot( (batas[-1], max), (0,0), warna, label=label ) 
```

```python
plot_linguistik( pelayanan['low'],  'r-', max=4 )
plot_linguistik( pelayanan['mid'],  'g-', max=4 )
plot_linguistik( pelayanan['high'], 'b-', max=4 )
plt.show()
```

```python
def membership_pelayanan_high(n):
  if n > 80: return 1
  elif n <= 60: return 0
  else: return (n - 60) / (80-60)
```

```python
def membership_pelayanan_mid(n):
  if n > 80 or n <= 20: return 0
  elif n > 20 and n <= 40: return (n - 20) / (40 - 20)
  elif n > 40 and n <= 60: return 1
  elif n > 60 and n <= 80: return (80 - n) / (80 - 60)
```

```python
def membership_pelayanan_low(n):
  if n <= 20: return 1
  elif n > 40: return 0
  else: return (40 - n) / (40 - 20)
```

```python
def membership_makanan_high(n):
  if n > 9: return 1
  if n <= 6: return 0
  return (n - 6) / (9 - 6)
```

```python
def membership_makanan_mid(n):
  if n > 8 or n <= 1: return 0
  if n > 1 and n <= 5: return (n - 1)/ (5 - 1)
  if n > 4 and n <= 6: return 1
  if n > 6 and n <= 9: return (9 - n) / (9 - 5)
```

```python
def membership_makanan_low(n):
  if n <= 1: return 1
  if n >= 5: return 0
  return (5 - n) / (5 - 1)
```

```python
fuzzy =[]

for i in range(len(data)):
  fuzzy.append([{'category': 'excellent', 'score':min(membership_makanan_high(data.loc[i]['makanan']), membership_pelayanan_high(data.loc[i]['pelayanan']))}, 
                {'category': 'excellent', 'score':min(membership_makanan_high(data.loc[i]['makanan']), membership_pelayanan_mid(data.loc[i]['pelayanan']))},
                {'category': 'average', 'score':min(membership_makanan_high(data.loc[i]['makanan']), membership_pelayanan_low(data.loc[i]['pelayanan']))},
                {'category': 'excellent', 'score':min(membership_makanan_mid(data.loc[i]['makanan']), membership_pelayanan_high(data.loc[i]['pelayanan']))},
                {'category': 'average', 'score':min(membership_makanan_mid(data.loc[i]['makanan']), membership_pelayanan_mid(data.loc[i]['pelayanan']))},
                {'category': 'average', 'score':min(membership_makanan_mid(data.loc[i]['makanan']), membership_pelayanan_low(data.loc[i]['pelayanan']))},
                {'category': 'average', 'score':min(membership_makanan_low(data.loc[i]['makanan']), membership_pelayanan_high(data.loc[i]['pelayanan']))},
                {'category': 'bad', 'score':min(membership_makanan_low(data.loc[i]['makanan']), membership_pelayanan_mid(data.loc[i]['pelayanan']))},
                {'category': 'bad', 'score':min(membership_makanan_low(data.loc[i]['makanan']), membership_pelayanan_low(data.loc[i]['pelayanan']))}])
fuzzy
```

```python
result = []
for i in range(len(fuzzy)):
    max_excellent = 0
    max_average = 0
    max_bad = 0
    for j in range(len(fuzzy[i])):
      if fuzzy[i][j]['category'] == 'excellent':
        max_excellent = max(max_excellent, fuzzy[i][j]['score'])
        #print(fuzzy[i][j]['score'], max_excellent)
      elif fuzzy[i][j]['category'] == 'average':
        max_average = max(max_average, fuzzy[i][j]['score'])
        #print(fuzzy[i][j]['score'], max_average)
      elif fuzzy[i][j]['category'] == 'bad':
        max_bad = max(max_bad, fuzzy[i][j]['score'])
        #print(fuzzy[i][j]['score'], max_bad)
    result.append([max_excellent, max_average, max_bad])
    

result
```
